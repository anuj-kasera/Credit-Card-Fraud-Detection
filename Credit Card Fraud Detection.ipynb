```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report, roc_auc_score
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc


def evaluate_model(model, X_data, y_data):
    # Predict on the data set
    y_pred = model.predict(X_data)
    
    # Confusion matrix
    confusion = confusion_matrix(y_data, y_pred)
    print("Confusion Matrix:")
    print(confusion)
    print("----"*15)
    
    # True positives, true negatives, false positives, false negatives
    TP = confusion[1, 1]
    TN = confusion[0, 0]
    FP = confusion[0, 1]
    FN = confusion[1, 0]
    
    # Accuracy
    accuracy = accuracy_score(y_data, y_pred)
    print("Accuracy:", accuracy)
    
    # Sensitivity (True Positive Rate)
    sensitivity = TP / float(TP + FN)
    print("Sensitivity (True Positive Rate):", sensitivity)
    
    # Specificity (True Negative Rate)
    specificity = TN / float(TN + FP)
    print("Specificity (True Negative Rate):", specificity)
    
    # F1 score
    f1 = f1_score(y_data, y_pred)
    print("F1-Score:", f1)
    print("----"*15)
    
    # Classification report
    print("Classification Report:")
    print(classification_report(y_data, y_pred))
    
    # Predicted probability for ROC AUC
    y_pred_proba = model.predict_proba(X_data)[:, 1]
    print("----"*15)
    
    # ROC AUC
    roc_auc = roc_auc_score(y_data, y_pred_proba)
    print("ROC AUC:", roc_auc)
```

# Undersampling

```python
# Importing undersampler library
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

# Instantiating the random undersampler
undersampler = RandomUnderSampler()

# Resampling X_train and y_train
X_train_undersampled, y_train_undersampled = undersampler.fit_resample(X_train, y_train)

print('Before sampling class distribution:', Counter(y_train))
print('New class distribution:', Counter(y_train_undersampled))
```

# Model building on balanced data with Undersampling

<h2> Logistic Regression

```python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression

# Creating KFold object with 5 splits
kf = KFold(n_splits=5, shuffle=True, random_state=4)

# Specify params
parameters = {"C": [0.01, 0.1, 1, 10, 100, 1000]}

# Specifying score as roc-auc
model_cv = GridSearchCV(estimator=LogisticRegression(),
                        param_grid=parameters, 
                        scoring='roc_auc', 
                        cv=kf, 
                        verbose=1,
                        return_train_score=True) 

# Fit the model
model_cv.fit(X_train_undersampled, y_train_undersampled)
```

```python
best_score = model_cv.best_score_
best_C = model_cv.best_params_['C']

print("The highest test roc_auc is {0} at C = {1}".format(best_score, best_C))
```

```python
# Logistic regression with optimal C

logistic_model = LogisticRegression(C=0.1)
logistic_model.fit(X_train_undersampled, y_train_undersampled)
```

<b> Prediction on Train set

```python
evaluate_model(logistic_model, X_train_undersampled, y_train_undersampled)
```

<b> Prediction on the test set

```python
evaluate_model(logistic_model, X_test, y_test)
```
